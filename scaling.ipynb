```python
import numpy as np
import matplotlib.pyplot as plt
```

## Strong scaling

Strong scaling on a grid of size 500x2500, starting with 1 processor up to 360 on 10 nodes.

```python
procs = np.array([1,2,4,8,12,18,24,36,2*36,3*36,4*36,6*36,8*36,10*36])
times1 = np.array([318.803,158.022,81.845,40.766,28.723,20.361,15.856,10.826,5.921,5.292,3.483,3.705,3.487,2.170])
times2 = np.array([317.471,158.551,80.016,40.766,28.606,20.708,15.657,10.826,5.874,4.425,3.483,2.998,2.563,2.170])
times = np.minimum(times1,times2)
speedup = times[0]/times
ideal_times = times[0] / procs

#labels = ['' for i in range(1,36,2)] + [1,2,3,4,6,8,10]
# Below 1 node ticks are 2 procs each, after that no minor ticks
#locs = list(range(1,36,2)) + [36,72,108,144,6*36,8*36,360]

labels = [1,10,100]
major = np.array(labels)
minor = np.array([i for i in range(2,10)] + [i for i in range(20,100,10)] + [i for i in range(200,1000,100)])

f, ax = plt.subplots(2,2)
f.set_size_inches(15,10)

ax0 = ax[0,0]
ax1 = ax[0,1]
ax2 = ax[1,0]
ax3 = ax[1,1]

ax0.plot(procs,times)
ax0.plot(procs,ideal_times)
ax0.set_xscale('log')
ax0.set_yscale('log')
ax0.set_title('Strong scaling - Runtime')
ax0.set_xlabel('CPUs')
ax0.set_ylabel('Time [s]')
ax0.set_xticks(major)
ax0.set_xticklabels(labels,minor=False)
ax0.set_xticks(minor,minor=True)
ax0.set_xticklabels([],minor=True)
ax0.set_yticks(major)
ax0.set_yticklabels(labels,minor=False)
#ax[0].tick_params(axis='x',which='minor',bottom='off')

ax1.plot(procs,times[0]/(procs*times))
ax1.plot(procs,procs/procs)
ax1.set_xscale('log')
ax1.set_title('Strong scaling - Efficiency')
ax1.set_xlabel('CPUs')
ax1.set_ylabel('Efficiency')
ax1.set_xticks(major)
ax1.set_xticklabels(labels,minor=False)
ax1.set_xticks(minor,minor=True)
ax1.set_xticklabels([],minor=True)

ax2.plot(procs,speedup)
ax2.plot(procs,ideal_times[0] / ideal_times)
#ax2.set_xscale('log')
#ax2.set_yscale('log')
ax2.set_title('Strong scaling - Speedup')
ax2.set_xlabel('CPUs')
ax2.set_ylabel('Speedup')
#ax2.set_xticks(major)
#ax2.set_xticklabels(labels,minor=False)
#ax2.set_xticks(minor,minor=True)
#ax2.set_xticklabels([],minor=True)
#ax2.set_yticks(major)
#ax2.set_yticklabels(labels,minor=False)

ax3.axis('off')
plt.show()
```

## Weak scaling

Weak scaling starting with a grid of size 100x500 on 1 processor, up to 100x180000 on 360 processors on 10 nodes. 

```python
procs = np.array([500,1000,2000,4000,6000,9000,12000,18000,36000,54000,72000,108000,144000,180000]) // 500
times1 = np.array([10.926,11.181,11.672,12.800,14.141,14.232,14.349,14.376,14.291,14.615,14.452,15.079,15.589,15.567])
times2 = np.array([11.002,11.273,11.705,12.421,14.141,14.194,14.257,14.433,14.892,14.554,14.504,14.665,14.905,14.796])
times = np.minimum(times1,times2)
speedup = times[0] / times * procs
ideal_times = times[0] * times / times 

labels = [1,10,100]
major = np.array(labels)
minor = np.array([i for i in range(2,10)] + [i for i in range(20,100,10)] + [i for i in range(200,1000,100)])

f, ax = plt.subplots(2,2)
f.set_size_inches(15,10)

ax0 = ax[0,0]
ax1 = ax[0,1]
ax2 = ax[1,0]
ax3 = ax[1,1]

ax0.plot(procs,times)
ax0.plot(procs,ideal_times)
ax0.set_xscale('log')
ax0.set_title('Weak scaling - Runtime')
ax0.set_xlabel('CPUs')
ax0.set_ylabel('Time [s]')
ax0.set_xticks(major)
ax0.set_xticklabels(labels,minor=False)
ax0.set_xticks(minor,minor=True)
ax0.set_xticklabels([],minor=True)

ax1.plot(procs,times[0]/times)
ax1.plot(procs,procs/procs)
ax1.set_xscale('log')
ax1.set_title('Weak scaling - Efficiency')
ax1.set_xlabel('CPUs')
ax1.set_ylabel('Efficiency')
ax1.set_xticks(major)
ax1.set_xticklabels(labels,minor=False)
ax1.set_xticks(minor,minor=True)
ax1.set_xticklabels([],minor=True)

ax2.plot(procs,speedup)
ax2.plot(procs,ideal_times[0] / ideal_times * procs)
ax2.set_title('Weak scaling - Speedup')
ax2.set_xlabel('CPUs')
ax2.set_ylabel('Speedup')

ax3.axis('off')
plt.show()
```
